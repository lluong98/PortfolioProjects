```python
# Looking at number of unique titles in the daily top 10

df['Title'].drop_duplicates().sort_values(ascending = False)
```

```python
#Looking for the titles in the Top 10 that are Netflix Exclusives

df.query('`Netflix Exclusive`== "Yes"')
```

```python
df[['Title','Netflix Exclusive', 'Type']].groupby(['Type','Netflix Exclusive']).count()
```

```python
#The longest amount of time each show or movie stayed in top 10.

sorted_df = df.sort_values(by=['Days In Top 10', 'Title'], ascending = [False, True]).drop_duplicates(subset = 'Title')
sorted_df
```

```python
#Table for the 10 shows that stayed in top 10 the longest

sorted_df.head(10)
```

```python
#Alternative method from using sort value function.

longest_top_10_shows = df.groupby(['Title', 'Type']).max().nlargest(10, 'Days In Top 10')
longest_top_10_shows
```

```python
type_table = df[['Title','Type']].groupby('Type').count()
type_table
```

```python
#Alternative to groupby

df['Type'].value_counts()
```

```python
df['Type'].value_counts().plot(kind = 'barh', yticks = [0,500,1000,2000,4000], rot = 0)
plt.ylabel("Types", labelpad = 20)
plt.xlabel("Number of Titles")
plt.title("Netflix Titles Broken into Types")
```

```python
#What is the average top 10 rank for each title?

avg_ranks = df[['Title','Rank']].groupby(['Title']).mean('Rank').sort_values(by = 'Rank', ascending = False)
avg_ranks
```

```python
#Let's see if there's any correlation between the numerical data

corr_matrix = df.corr()
corr_matrix
```

```python
#Visual representation of correlation matrix

sns.heatmap(corr_matrix, annot = True)

plt.title('Pearson Correlation Matrix for Numeric Features')
plt.show()
```
